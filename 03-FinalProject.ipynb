## Research Question

Between 2006 and 2015, do North America, Europe, and Japan differ in their market-share distribution across four genre categories (Action/Competitive, Narrative, Strategy/Cognitive, and Misc) on the PS3, X360, Wii, DS, PSP, and PC platforms, and have these regional differences changed over time?

We use a video game sales dataset containing regional sales information for North America, Europe, and Japan. To focus on a consistent multi-platform generation era and avoid lifecycle distortions across console generations, we restrict the analysis to the years 2006–2015 and to the major overlapping platforms PS3, X360, Wii, DS, PSP, and PC.

We group the original 12 genres into four broader genre categories (Action/Competitive, Narrative, Strategy/Cognitive, and Misc) to reduce dimensionality while preserving interpretability. For each region, year, and platform, we compute genre market share as genre-specific regional sales divided by total regional sales for that platform-year.

We then compare genre-share distributions across regions using statistical inference tests and estimate regression models to evaluate whether the magnitude of regional differences changes over time.

## Background and Prior Work

The video game industry is a global market in which games are produced, distributed, and consumed across diverse cultural and economic contexts. A central question in this globalized environment is whether player preferences are converging worldwide or whether they remain shaped by regional tastes, platform ecosystems, and local industry structures. As the industry has expanded, scholars have debated whether players in different regions increasingly gravitate toward similar genres and platforms or whether long‑standing regional identities continue to generate distinct patterns of demand. This tension between global convergence and regional differentiation motivates a closer empirical examination of how video game purchases vary across regions and how those patterns evolve over time.

Prior work has demonstrated that commercial performance varies widely across titles, even within the same release year. Researchers have identified product characteristics, especially platform and genre, as major determinants of sales outcomes. Studies using global sales rankings demonstrate that both platform and genre exert significant influence on units sold, suggesting that hardware ecosystems and content categories shape how games compete in the marketplace (Wang, 2025). In addition, regional sales variables explain further variation in performance, indicating that demand conditions differ meaningfully across geographic markets. These findings motivate a region‑sensitive approach in which regional market characteristics are treated as integral components of explanatory models rather than secondary factors.

Additionally, regional variation in video game consumption is not only cross‑sectional, but also temporally persistent. Studies of long run sales patterns have found that, while some genres attain widespread popularity, regional markets still tend to maintain distinct preferences that reflect cultural tastes and local industry structures (Zhang, 2021). This suggests that differences in the composition of regional game libraries are the result of relatively stable underlying demand conditions, rather than short‑term trends or one‑off blockbuster releases. In other words, regions do not simply converge on the same set of titles. Instead, they continue to favor different mixes of genres and platforms over time, indicating that regional identity and market context play a lasting role in shaping what players choose to buy.

“Predictive Analytics for Video Game Sales” by Rifaz primarily focuses on predictive modeling of sales outcomes rather than on comparative analysis of regional preferences. This prior work shows that global sales outcomes can be predicted with very high accuracy using regional sales data, indicating that regional markets exhibit stable and structured sales patterns that consistently explain global success. However, it treats regions mainly as inputs to prediction rather than as objects of comparison. In particular, it does not examine whether regions systematically differ in their genre preferences after accounting for total market size, nor does it analyze how those differences change over time. This project builds on that analytical foundation but reframes the problem toward comparative and longitudinal analysis, focusing on genre market share by region and year rather than on "The impact of platform on global video game sales" has identified gaming platforms as a key structural factor influencing global video game sales patterns.

Their study showed that Nintendo platforms—particularly the Wii and Nintendo DS—consistently outperformed competing consoles globally, while Microsoft and Sony platforms occupied lower sales tiers. These findings suggest that platform strategy, vertical integration, and regional platform adoption play a central role in shaping observed sales distributions. This work provides an important foundation for our analysis of regional sales differences, particularly in understanding why Japan’s sales distribution diverges from North America and Europe, where non-Nintendo platforms are more dominant.

Taken together, the literature suggests that platform, genre, and region all play important roles in shaping video game consumption, but it leaves open key questions about how regional genre preferences differ, how these differences vary across platforms, and whether they change over time. ## Regional Genre Preferences: Year-by_year Market Share (2006 - 2015)

We examine how genre market share evolved annually within each region. By calculating market share per year, we can observe temporal shifts in preferences rather than relying on aggregate summaries that may ignore important trends.

```python
platforms = ['PS3', 'X360', 'Wii', 'DS', 'PSP', 'PC']
df_pref = df[(df['Year'].between(2006, 2015)) & (df['Platform'].isin(platforms))].copy()

# Calculate market share by year, region, and genre
pref_long = df_pref.melt(
    id_vars=['Year', 'Genre_Category'],
    value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales'],
    var_name='Region',
    value_name='Sales_Millions'
)
pref_long['Region'] = pref_long['Region'].str.replace('_Sales', '', regex=False)

# Aggregate by Year, Region, Genre
yearly_genre = (
    pref_long.groupby(['Year', 'Region', 'Genre_Category'], as_index=False)['Sales_Millions']
    .sum()
)

# Calculate market share within each year-region
yearly_genre['Market_Share_Pct'] = (
    yearly_genre.groupby(['Year', 'Region'])['Sales_Millions']
    .transform(lambda x: x / x.sum() * 100)
)
```

We now show NA, EU, and JP sales by year and genre category in separate tables below:

```python
region_data = yearly_genre[yearly_genre['Region'] == 'NA']
pivot_table_na = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

print("Pivot Table for NA Region:")
pivot_table_na
```

```python
region_data = yearly_genre[yearly_genre['Region'] == 'EU']
pivot_table_eu = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

print("Pivot Table for EU Region:")
pivot_table_eu
```

```python
region_data = yearly_genre[yearly_genre['Region'] == 'JP']
pivot_table_jp = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

print("Pivot Table for JP Region:")
pivot_table_jp
```

Now we make visualization of the genre market share within each region by year.

```python
# Visualize for all regions together
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

for i, region in enumerate(region_order):
    ax = axes[i]
    region_data = yearly_genre[yearly_genre['Region'] == region]
    
    # Pivot for stacked area plot
    pivot = region_data.pivot(index='Year', columns='Genre_Category', values='Market_Share_Pct')[genre_order]
    
    # Stacked area chart
    ax.stackplot(
        pivot.index, 
        pivot['Action/Competitive'],
        pivot['Narrative'],
        pivot['Strategy/Cognitive'],
        pivot['Misc'],
        labels=genre_order,
        alpha=0.8
    )
    
    ax.set_title(f'{region}', fontweight='bold', fontsize=11)
    ax.set_xlabel('Year')
    if i == 0:
        ax.set_ylabel('Market Share (%)')
    ax.set_ylim(0, 100)
    ax.grid(True, alpha=0.2, axis='y')
    ax.legend(loc='upper left', fontsize=8, frameon=False)

plt.tight_layout()
plt.suptitle('Genre Market Share Composition by Region (2006–2015)', y=1.02, fontweight='bold', fontsize=12)
plt.show()
```

#### Load Processed Data & Compute Market Shares

```python
df_vg = pd.read_csv('data/02-processed/videogamesales_clean.csv')

PLATFORMS = ['PS3', 'X360', 'Wii', 'DS', 'PSP', 'PC']
df_vg = df_vg[
    df_vg['Platform'].isin(PLATFORMS) &
    df_vg['Year'].between(2006, 2015)
].copy()
df_vg['Genre_Cat'] = df_vg['Genre'].map(GENRE_MAP)

print(f'Loaded: {df_vg.shape[0]:,} rows | Platforms: {sorted(df_vg["Platform"].unique())} | Years: {df_vg["Year"].min()}–{df_vg["Year"].max()}')
print(df_vg['Genre_Cat'].value_counts())
```

```python
# Aggregate sales by Platform x Year x Genre_Cat
agg_vg = df_vg.groupby(['Platform', 'Year', 'Genre_Cat'], as_index=False).agg(
    NA_Sales=('NA_Sales', 'sum'),
    EU_Sales=('EU_Sales', 'sum'),
    JP_Sales=('JP_Sales', 'sum')
)

# Genre market share = genre-specific regional sales / total regional sales for that Platform-Year
totals = agg_vg.groupby(['Platform', 'Year'])[['NA_Sales','EU_Sales','JP_Sales']].transform('sum')
agg_vg['NA_Share'] = agg_vg['NA_Sales'] / totals['NA_Sales']
agg_vg['EU_Share'] = agg_vg['EU_Sales'] / totals['EU_Sales']
agg_vg['JP_Share'] = agg_vg['JP_Sales'] / totals['JP_Sales']

regions = {'NA': 'NA_Share', 'EU': 'EU_Share', 'JP': 'JP_Share'}

print('Market share table sample (first 12 rows):')
agg_vg.head(12)
```

#### 1.2 Chi-Square Test - Do Genre Distributions Differ Across Regions?

We build a contingency table of total sales per region x genre category and test whether the genre distribution is independent of region.

- H0: The distribution of sales across genre categories is the same in NA, EU, and JP
- H1: At least one region has a significantly different genre distribution

```python
contingency = df_vg.groupby('Genre_Cat')[['NA_Sales','EU_Sales','JP_Sales']].sum()
contingency.columns = ['NA', 'EU', 'JP']
contingency = contingency.reindex(GENRES)

print('Contingency table (total sales in millions):')
print(contingency.round(2))
print()

chi2_vg, p_vg, dof_vg, expected_vg = chi2_contingency(contingency)
n_vg = contingency.values.sum()
cramers_v_vg = np.sqrt(chi2_vg / (n_vg * (min(contingency.shape) - 1)))
effect_vg = 'negligible' if cramers_v_vg < 0.1 else 'small' if cramers_v_vg < 0.3 else 'medium' if cramers_v_vg < 0.5 else 'large'

print(f'Chi-Square Statistic : {chi2_vg:.4f}')
print(f'Degrees of Freedom   : {dof_vg}')
print(f'p-value              : {p_vg:.2e}')
print(f"Cramer's V           : {cramers_v_vg:.4f}  ({effect_vg} effect size)")
print()
print('Reject H0 - genre distributions differ significantly across regions (p < 0.05).' if p_vg < 0.05
      else 'Fail to reject H0 - no significant difference detected.')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(len(GENRES))
w = 0.25
for i, (region, color) in enumerate(REGION_COLORS.items()):
    ax.bar(x + i*w, contingency[region].values, width=w, label=region, color=color, edgecolor='white')
ax.set_xticks(x + w)
ax.set_xticklabels(GENRES, fontsize=10)
ax.set_ylabel('Total Sales (millions)')
ax.set_title(
    f'Total Sales by Genre Category and Region (2006-2015)\n'
    f'Chi2 = {chi2_vg:.2f},  p = {p_vg:.2e},  Cramers V = {cramers_v_vg:.3f} ({effect_vg} effect)'
)
ax.legend(title='Region')
plt.tight_layout()
plt.show()
```

#### 1.3 Kruskal-Wallis Test + Mann-Whitney Post-Hoc

The Chi-Square test uses total sales aggregates. Here, we use the distribution of per-(Platform, Year) market-share values to test whether regional differences hold at finer granularity.

- Kruskal-Wallis tests whether at least one region differs per genre (non-parametric, no normality assumption)
- Mann-Whitney U pairwise post-hoc tests identity which region pairs drive the difference, with Bonferroni correction for multiple comparisons

```python
print('Kruskal-Wallis H-Test: Regional Genre Market Share Distributions')
print('=' * 65)
kruskal_results = {}
for genre in GENRES:
    sub = agg_vg[agg_vg['Genre_Cat'] == genre]
    na = sub['NA_Share'].dropna()
    eu = sub['EU_Share'].dropna()
    jp = sub['JP_Share'].dropna()
    h, p = kruskal(na, eu, jp)
    kruskal_results[genre] = {'H': h, 'p': p}
    sig = 'Significant' if p < 0.05 else 'Not significant'
    print(f'{genre:<25}  H = {h:7.3f}  p = {p:.4f}  -> {sig}')

print()
print('p < 0.05 means at least one region differs significantly in that genre market share.')
```

```python
region_pairs = list(combinations(regions.items(), 2))
adj_alpha = 0.05 / (len(GENRES) * len(region_pairs))

print(f'Mann-Whitney U Post-Hoc Tests (Bonferroni-corrected alpha = {adj_alpha:.4f})')
print('=' * 72)

for genre in GENRES:
    sub = agg_vg[agg_vg['Genre_Cat'] == genre]
    print(f' {genre}')
    for (r1name, r1col), (r2name, r2col) in region_pairs:
        u, p = mannwhitneyu(sub[r1col].dropna(), sub[r2col].dropna(), alternative='two-sided')
        sig = '[SIG]' if p < adj_alpha else '      '
        print(f'     {r1name} vs {r2name}  U = {u:6.0f}  p = {p:.4f}   {sig}')
    print()
```

#### 1.4 Monte Carlo Simulation - Are Regional Differences Due to Chance?

We test the null hypothesis (no region differences) by randomly permuting sales across regions 10,000 times per genre. The observed variance across regional mean shares is compared to the permuted null distribution to compute an empirical p-value

```python
np.random.seed(42)
N_SIM = 10_000
mc_results = {}

for genre in GENRES:
    sub = agg_vg[agg_vg['Genre_Cat'] == genre].copy()
    na = sub['NA_Share'].dropna().values
    eu = sub['EU_Share'].dropna().values
    jp = sub['JP_Share'].dropna().values
    obs_means = np.array([na.mean(), eu.mean(), jp.mean()])
    obs_var   = np.var(obs_means)
    all_shares = np.concatenate([na, eu, jp])
    n = len(all_shares) // 3
    sim_vars = []
    for _ in range(N_SIM):
        s = np.random.choice(all_shares, size=3*n, replace=True)
        sim_vars.append(np.var([s[:n].mean(), s[n:2*n].mean(), s[2*n:].mean()]))
    sim_vars = np.array(sim_vars)
    p_mc = (sim_vars >= obs_var).mean()
    mc_results[genre] = {'obs_var': obs_var, 'sim_vars': sim_vars, 'p_mc': p_mc}

print('Monte Carlo Results (10,000 bootstrap simulations per genre)')
print('=' * 68)
for genre, res in mc_results.items():
    sig = 'Significant' if res['p_mc'] < 0.05 else 'Not significant'
    print(f'{genre:<25}  obs_var = {res["obs_var"]:.6f}   MC p = {res["p_mc"]:.4f}   -> {sig}')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
for ax, (genre, res), color in zip(axes.flatten(), mc_results.items(), GENRE_COLORS):
    ax.hist(res['sim_vars'], bins=60, color=color, alpha=0.7, edgecolor='white', label='Simulated null')
    ax.axvline(res['obs_var'], color='black', linewidth=2, linestyle='--',
               label=f'Observed  (MC p = {res["p_mc"]:.4f})')
    ax.set_title(genre)
    ax.set_xlabel('Variance of regional mean shares')
    ax.set_ylabel('Frequency')
    ax.legend(fontsize=8)
plt.suptitle('Monte Carlo Simulation: Observed vs Null Distribution of Regional Variance (10,000 sims)', fontsize=13)
plt.tight_layout()
plt.show()
```

#### 1.5 OLS Regression - Has the Regional Gap Changed Over Time?

The EDA showed visually that regional differences appear stable over 2006-2015. Here we formally test this by regressing the maximum pairwise regional gap in genre share on year.

- HO (our hypothesis): Slope = 0 - the region gap has not changed over time
- Slope > 0: gap is widening (regions diverging)
- Slope < 0: gap is narrowing (regions converging)
- p > 0.05: consistent with the hypothesis that the gap is unchanged

```python
# Annual mean genre share per region (year-weighted, not sales-weighted)
yearly = agg_vg.groupby(['Year', 'Genre_Cat']).agg(
    NA=('NA_Share', 'mean'),
    EU=('EU_Share', 'mean'),
    JP=('JP_Share', 'mean')
).reset_index()

# Max pairwise absolute gap across 3 regions per year-genre
yearly['gap_NA_EU'] = (yearly['NA'] - yearly['EU']).abs()
yearly['gap_NA_JP'] = (yearly['NA'] - yearly['JP']).abs()
yearly['gap_EU_JP'] = (yearly['EU'] - yearly['JP']).abs()
yearly['max_gap']   = yearly[['gap_NA_EU', 'gap_NA_JP', 'gap_EU_JP']].max(axis=1)

print('OLS Regression: Max Regional Genre Share Gap ~ Year')
print('=' * 72)
ols_results = {}
for genre in GENRES:
    sub = yearly[yearly['Genre_Cat'] == genre].sort_values('Year')
    slope, intercept, r, p, se = stats.linregress(sub['Year'].values, sub['max_gap'].values)
    ols_results[genre] = {
        'slope': slope, 'intercept': intercept,
        'r2': r**2, 'p': p,
        'x': sub['Year'].values, 'y': sub['max_gap'].values
    }
    trend = 'widening' if slope > 0 else 'narrowing'
    sig = '(p < 0.05) gap IS changing' if p < 0.05 else '(p > 0.05) gap unchanged'
    print(f'{genre:<25}  slope = {slope:+.5f}   R2 = {r**2:.3f}   p = {p:.4f}   -> {trend}  {sig}')

print()
print('Hypothesis predicts p > 0.05 (no significant change) for all genres.')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
for ax, (genre, res), color in zip(axes.flatten(), ols_results.items(), GENRE_COLORS):
    x, y = res['x'], res['y']
    ax.scatter(x, y, color=color, s=60, zorder=3, label='Observed max gap')
    ax.plot(x, res['slope']*x + res['intercept'], color='black', linewidth=1.8,
            linestyle='--', label=f'OLS  slope = {res["slope"]:+.5f}')    
    ax.set_title(genre)
    ax.set_xlabel('Year')
    ax.set_ylabel('Max regional gap in share')
    ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
    ax.legend(fontsize=8)
    ax.text(0.05, 0.92, f'R2 = {res["r2"]:.3f} p = {res["p"]:.3f}', transform=ax.transAxes, fontsize=8, color='grey')

plt.suptitle('OLS Regression: Regional Genre Share Gap Over Time (2006-2015)', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
df_steam = pd.read_csv('data/02-processed/steamgames_clean.csv')
df_steam = df_steam[df_steam['Year'].between(2006, 2015)].copy()

# Parse ownership ranges to numeric midpoint
def parse_owners(s):
    try:
        lo, hi = [int(x.strip().replace(',', '')) for x in str(s).split('-')]
        return (lo + hi) / 2
    except:
        return np.nan

df_steam['Owners_Mid'] = df_steam['Estimated owners'].apply(parse_owners)

# Ordinal ownership tier
OWNER_ORDER = ['0 - 20000', '20000 - 50000', '50000 - 100000', '100000 - 200000', 
               '200000 - 500000', '500000 - 1000000', '1000000 - 2000000', 
               '2000000 - 5000000', '5000000 - 10000000', '10000000 - 20000000', 
               '20000000 - 50000000', '50000000 - 100000000', '100000000 - 200000000']
df_steam['Owner_Tier'] = pd.Categorical(df_steam['Estimated owners'], categories=OWNER_ORDER, ordered=True)
df_steam['Owner_Rank'] = df_steam['Owner_Tier'].cat.codes

# Map primary genre to the same 4 categories used in Part 1
def map_steam_genre(g):
    g = str(g).split(',')[0].strip()
    mapping = {
        'Action': 'Action/Competitive', 'Sports': 'Action/Competitive',
        'Racing': 'Action/Competitive', 'Fighting': 'Action/Competitive',
        'Adventure': 'Narrative', 'RPG': 'Narrative',
        'Strategy': 'Strategy/Cognitive', 'Puzzle': 'Strategy/Cognitive',
        'Simulation': 'Strategy/Cognitive',
        'Indie': 'Misc', 'Casual': 'Misc', 'Free To Play': 'Misc',
        'Early Access': 'Misc', 'Massively Multiplayer': 'Misc'
    }
    return mapping.get(g, 'Misc')

df_steam['Genre_Cat'] = df_steam['Genres'].apply(map_steam_genre)

print(f'Steam -- {df_steam.shape[0]:,} rows (2006-2015)')
print('Genre category distribution:')
print(df_steam['Genre_Cat'].value_counts())
df_steam.head()
    
```

```python
df_gp = pd.read_csv('data/02-processed/googleplaystore_clean.csv')

def parse_installs(s):
    s = str(s).replace(',', '').replace('+', '').strip()
    try:
        return float(s)
    except:
        return np.nan

df_gp['Installs_Num'] = df_gp['Installs'].apply(parse_installs)
df_gp = df_gp.dropna(subset=['Installs_Num'])

install_order = ['1+', '5+', '10+', '50+', '100+', '500+', '1,000+', '5,000+', '10,000+',
                 '50,000+', '100,000+', '500,000+', '1,000,000+', '5,000,000+', '10,000,000+',
                 '50,000,000+', '100,000,000+', '500,000,000+', '1,000,000,000+']
df_gp['Install_Tier'] = pd.Categorical(df_gp['Installs'], categories=install_order, ordered=True)
df_gp['Install_Rank'] = df_gp['Install_Tier'].cat.codes

def map_gp_genre(g):
    g = str(g).split(';')[0].strip()
    mapping = {
        'Action': 'Action/Competitive', 'Action & Adventure': 'Action/Competitive',
        'Sports': 'Action/Competitive', 'Racing': 'Action/Competitive', 'Adventure': 'Narrative',
        'Role Playing': 'Narrative', 'Strategy': 'Strategy/Cognitive', 'Simulation': 'Strategy/Cognitive',
        'Puzzle': 'Strategy/Cognitive', 'Brain Games': 'Strategy/Cognitive', 'Arcade': 'Misc', 'Casual': 'Misc',
        'Card': 'Misc', 'Board': 'Misc', 'Music': 'Misc', 'Trivia': 'Misc', 'Word': 'Misc'
    }
    return mapping.get(g, 'Misc')

df_gp['Genre_Cat'] = df_gp['Genres'].apply(map_gp_genre)

print(f'Google Play -- {df_gp.shape[0]:,} rows')
print('Genre category distribution:')
print(df_gp['Genre_Cat'].value_counts())
df_gp.head()
```

## Discusison and Conclusion

This project evaluated whether Noth American, Europe, and Japan differ in their preferences in video game genres across major platforms during the 2006 - 2015 console generation. We also examined whether the relative sales gap between these regions changed over time. By integrating three datasets and conducting tempral, reigonal, and platform level analyses, we found strong and consistent evidence that regional genre preferences and distinct and stable, consistent with the findings of Wang (2025) & Zhang (2021).

Our primary analysis revealed clear and persistent differences in genre composition across regions. We found that North America and Europe exhibit nearly identical market structures, with the Action/Competitive genre titles being the most dominant in annual market share throughout the decade (55-70%). These findings are consistent with the findings of Zhang (2021), who identified Western players as disproportionately drawn to high intensity games in the action genre

However, narrative titles have a much larger share in Japan, often being 25-50%. This demonstrates a contrast to NA and EU, reflecting Japan;s long standing affinity for story driven and role playing games. Although Action/Competitive titles have shown growth in Japan over time, the region maintains a more balanced distribution across genres. These results support the hypothesis that **regional genre preferences reflect enduring cultural and market differences rather than short term fluctations.**

Temporal analysis further demonstrated that although total sales peaked around 2008 with a decline after, the relative ranking of regions currently remains stable. North American consistently accounts for the largest share of global sales, followed by Europe and Japan. This stability is consistent with Wang (2025), who observed that NA and EU dominate global sales.

Platform level analysis further supports these conclusions. Even after conditioning on platform, regional genre patterns persist. For example on PS3 and X360, Action/Competitive titles still dominated in NA and EU, with Japan still maintaining a nigher Narrative share. The results suggest that regional differences are not artifacts of platform mix, but reflect genuine differences in consumer preference.

Some inconsistencies emerged when comparing across datasets. The Steam Dataset (Dataset 2) suggested genre distributions that did not always align with console based patterns. This is likely because of the global nature of PC digital distribution, or the categorical and range based nature of estimated owners. These constraints make the Steam results slightly less reliable for regional inference. Additionally, Dataset 3 lacked both temporal and regional information, preventing a more meaningful comparison.

There were several limitations that affect the generalizability of our findings. Firstly, there is retail bias in Dataset 1. Digital sales, especially after 2010 are underepresented. Another limitation is the market evolution post 2015. The rise of more digital distribution and mobile dominance could change the patterns that we discovered. Despite these limitations, the consistency of results across multiple analytical approaches strengthens confidence in our findings.

If this research were to continue, we would suggest incorporating digital only sales datasets in order to correct retail bias. Another interesting topic would be examining player engagement metrics in order to complement sales based measures. For example, we could look at average playtime.

Taken together, our analyses demonstrated that regional differences in video game genre preferences are persistent, robust, and grounded in our culture. North America and Europe form a cohesive Western market with convergent tastes, while Japan has a distinct preference structured around narrative and role playing games.
