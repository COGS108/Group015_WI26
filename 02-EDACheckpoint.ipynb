# Research Question

Between 2006 and 2015, do North America, Europe, and Japan differ in their market-share distribution across four genre categories (Action/Competitive, Narrative, Strategy/Cognitive, and Misc) on the PS3, X360, Wii, DS, PSP, and PC platforms, and have these regional differences changed over time?

We use a video game sales dataset containing regional sales information for North America, Europe, and Japan. To focus on a consistent multi-platform generation era and avoid lifecycle distortions across console generations, we restrict the analysis to the years 2006–2015 and to the major overlapping platforms PS3, X360, Wii, DS, PSP, and PC.

We group the original 12 genres into four broader genre categories (Action/Competitive, Narrative, Strategy/Cognitive, and Misc) to reduce dimensionality while preserving interpretability. For each region, year, and platform, we compute genre market share as genre-specific regional sales divided by total regional sales for that platform-year.

We then compare genre-share distributions across regions using statistical inference tests and estimate regression models to evaluate whether the magnitude of regional differences changes over time.


## Background and Prior Work

The video game industry is a global market in which games are produced, distributed, and consumed across diverse cultural and economic contexts. A central question in this globalized environment is whether player preferences are converging worldwide or whether they remain shaped by regional tastes, platform ecosystems, and local industry structures. As the industry has expanded, scholars have debated whether players in different regions increasingly gravitate toward similar genres and platforms or whether long‑standing regional identities continue to generate distinct patterns of demand. This tension between global convergence and regional differentiation motivates a closer empirical examination of how video game purchases vary across regions and how those patterns evolve over time.

Prior work has demonstrated that commercial performance varies widely across titles, even within the same release year. Researchers have identified product characteristics, especially platform and genre, as major determinants of sales outcomes. Studies using global sales rankings demonstrate that both platform and genre exert significant influence on units sold, suggesting that hardware ecosystems and content categories shape how games compete in the marketplace (Wang, 2025). In addition, regional sales variables explain further variation in performance, indicating that demand conditions differ meaningfully across geographic markets. These findings motivate a region‑sensitive approach in which regional market characteristics are treated as integral components of explanatory models rather than secondary factors.

Additionally, regional variation in video game consumption is not only cross‑sectional, but also temporally persistent. Studies of long run sales patterns have found that, while some genres attain widespread popularity, regional markets still tend to maintain distinct preferences that reflect cultural tastes and local industry structures (Zhang, 2021). This suggests that differences in the composition of regional game libraries are the result of relatively stable underlying demand conditions, rather than short‑term trends or one‑off blockbuster releases. In other words, regions do not simply converge on the same set of titles. Instead, they continue to favor different mixes of genres and platforms over time, indicating that regional identity and market context play a lasting role in shaping what players choose to buy.

“Predictive Analytics for Video Game Sales” by Rifaz primarily focuses on predictive modeling of sales outcomes rather than on comparative analysis of regional preferences. This prior work shows that global sales outcomes can be predicted with very high accuracy using regional sales data, indicating that regional markets exhibit stable and structured sales patterns that consistently explain global success. However, it treats regions mainly as inputs to prediction rather than as objects of comparison. In particular, it does not examine whether regions systematically differ in their genre preferences after accounting for total market size, nor does it analyze how those differences change over time. This project builds on that analytical foundation but reframes the problem toward comparative and longitudinal analysis, focusing on genre market share by region and year rather than on "The impact of platform on global video game sales" has identified gaming platforms as a key structural factor influencing global video game sales patterns. 

Babb, Terry, and Dana (2013) examined worldwide video game sales across major platforms between 2006 and 2011 and found significant differences in sales performance attributable to platform ecosystems rather than individual game attributes alone. Their study showed that Nintendo platforms—particularly the Wii and Nintendo DS—consistently outperformed competing consoles globally, while Microsoft and Sony platforms occupied lower sales tiers. These findings suggest that platform strategy, vertical integration, and regional platform adoption play a central role in shaping observed sales distributions. This work provides an important foundation for our analysis of regional sales differences, particularly in understanding why Japan’s sales distribution diverges from North America and Europe, where non-Nintendo platforms are more dominant.

Taken together, the literature suggests that platform, genre, and region all play important roles in shaping video game consumption, but it leaves open key questions about how regional genre preferences differ, how these differences vary across platforms, and whether they change over time. #### EDA Setup: Load Processed Data and Define Genre Groups

In the next code cell, we initialize the analysis dataset and feature definitions used throughout the EDA.

- We load the already-cleaned dataset from `data/02-processed/videogamesales_clean.csv` so the analysis is reproducible without re-running earlier wrangling steps.
- We map the original 12 game genres into 4 broader categories (`Action/Competitive`, `Narrative`, `Strategy/Cognitive`, `Misc`) to reduce dimensionality while preserving interpretability.
- We also restrict to the project scope (years 2006–2015 and the six core platforms: PS3, X360, Wii, DS, PSP, PC), then define consistent genre/region orderings for downstream plots and tests.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data/02-processed/videogamesales_clean.csv')

# Mapping 12 genres into 4 broader categories
genre_mapping = {
    'Action': 'Action/Competitive', 'Shooter': 'Action/Competitive',
    'Fighting': 'Action/Competitive', 'Racing': 'Action/Competitive',
    'Sports': 'Action/Competitive', 'Role-Playing': 'Narrative',
    'Adventure': 'Narrative', 'Strategy': 'Strategy/Cognitive',
    'Simulation': 'Strategy/Cognitive', 'Puzzle': 'Strategy/Cognitive',
    'Misc': 'Misc', 'Platform': 'Misc'
}
df['Genre_Category'] = df['Genre'].map(genre_mapping)

# Filter the 2006-2015, 6 platforms
platforms = ['PS3', 'X360', 'Wii', 'DS', 'PSP', 'PC']
df_filtered = df[
    (df['Year'].between(2006, 2015, inclusive='both')) &
    (df['Platform'].isin(platforms))
].copy()

# Genre ordering for consistent visualization
genre_order = ['Action/Competitive', 'Narrative', 'Strategy/Cognitive', 'Misc']
region_order = ['NA', 'EU', 'JP']

# Showing first few rows of filtered data
df_filtered[['Name', 'Platform', 'Year', 'Genre', 'Genre_Category', 'NA_Sales', 'EU_Sales', 'JP_Sales']].head(10)
```

## Regional Market Sizes (2006-2015)

We first establish the relative market sizes across North America, Europe, and Japan to contextualize the genre preference analysis that follows.

```python
regional_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales']
market_share_table = pd.DataFrame({
    'total_sales_millions': df_filtered[regional_cols].sum(),
    'market_share_pct': (df_filtered[regional_cols].sum() / df_filtered[regional_cols].sum().sum() * 100)
}).round(2)
market_share_table
```

```python
yearly_totals = df_filtered.groupby('Year')[regional_cols].sum()
yearly_totals.plot(kind='line', figsize=(10, 6))
plt.title('Total Regional Sales Over Time (2006–2015)', fontweight='bold', fontsize=12)
plt.xlabel('Year')
plt.ylabel('Total Sales (millions)')
plt.legend(title='Region', frameon=False)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

The temporal plot shows that total sales in all three regions peak around 2008–2009 and decline thereafter, consistent with the late lifecycle of the seventh console generation. While the overall market contracts after this period, the relative ranking of regions remains consistent throughout the decade.

Across 2006–2015, North America accounts for the largest share of total sales (55.4%), followed by Europe (33.9%) and Japan (10.7%). Although absolute sales decline in all regions after 2009, their proportional contributions to total sales remain broadly stable over time.

Because the regions differ substantially in market size, we analyze genre preferences using market share, defined as the proportion of total regional sales attributable to each genre in a given year. This normalization removes scale effects and allows meaningful comparison of demand composition across regions, which directly supports our research question regarding regional genre preference patterns.

## Regional Genre Preferences: Year-by-Year Market Share (2006–2015)

We examine how genre market share evolved annually within each region. By calculating market share per year, we can observe temporal shifts in preferences rather than relying on aggregate summaries that may obscure important trends.

```python
platforms = ['PS3', 'X360', 'Wii', 'DS', 'PSP', 'PC']
df_pref = df[(df['Year'].between(2006, 2015)) & (df['Platform'].isin(platforms))].copy()

# Calculate market share by year, region, and genre
pref_long = df_pref.melt(
    id_vars=['Year', 'Genre_Category'],
    value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales'],
    var_name='Region',
    value_name='Sales_Millions'
)
pref_long['Region'] = pref_long['Region'].str.replace('_Sales', '', regex=False)

# Aggregate by Year, Region, Genre
yearly_genre = (
    pref_long.groupby(['Year', 'Region', 'Genre_Category'], as_index=False)['Sales_Millions']
    .sum()
)

# Calculate market share within each year-region
yearly_genre['Market_Share_Pct'] = (
    yearly_genre.groupby(['Year', 'Region'])['Sales_Millions']
    .transform(lambda x: x / x.sum() * 100)
)
```

We now show NA, EU, and JP sales by year and genre category in separate tables below:

```python
region_data = yearly_genre[yearly_genre['Region'] == 'NA']
pivot_table_na = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

pivot_table_na
```

```python
region_data = yearly_genre[yearly_genre['Region'] == 'EU']
pivot_table_eu = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

pivot_table_eu
```

```python
region_data = yearly_genre[yearly_genre['Region'] == 'JP']
pivot_table_jp = region_data.pivot(
    index='Year', 
    columns='Genre_Category', 
    values='Market_Share_Pct'
)[genre_order].round(1)

pivot_table_jp
```

```python
# Visualize for all regions together
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

for i, region in enumerate(region_order):
    ax = axes[i]
    region_data = yearly_genre[yearly_genre['Region'] == region]
    
    # Pivot for stacked area plot
    pivot = region_data.pivot(index='Year', columns='Genre_Category', values='Market_Share_Pct')[genre_order]
    
    # Stacked area chart
    ax.stackplot(
        pivot.index, 
        pivot['Action/Competitive'],
        pivot['Narrative'],
        pivot['Strategy/Cognitive'],
        pivot['Misc'],
        labels=genre_order,
        alpha=0.8
    )
    
    ax.set_title(f'{region}', fontweight='bold', fontsize=11)
    ax.set_xlabel('Year')
    if i == 0:
        ax.set_ylabel('Market Share (%)')
    ax.set_ylim(0, 100)
    ax.grid(True, alpha=0.2, axis='y')
    ax.legend(loc='upper left', fontsize=8, frameon=False)

plt.tight_layout()
plt.suptitle('Genre Market Share Composition by Region (2006–2015)', y=1.02, fontweight='bold', fontsize=12)
plt.show()
```

## Platform-Level Analysis: Genre Market Share by Region Across All Six Platforms

We examine regional genre preferences **within each platform** to determine whether regional differences persist when accounting for platform.

### Analytical Approach

1. **Compute market share** within each (Year, Platform, Region) combination
2. **Compute the mean of yearly market shares (2006–2015)** for each (Platform, Region, Genre). These values represent the unweighted average of annual market shares rather than sales-weighted totals across the entire period.
3. **Visualize all platforms simultaneously** using faceted bar charts

This symmetric approach treats all six platforms equally and allows direct comparison of regional genre composition within each hardware ecosystem. By conditioning on platform, we evaluate whether regional differences remain even when platform adoption patterns are held constant.

```python
pref_platform = df_pref.melt(
    id_vars=['Year', 'Platform', 'Genre_Category'],
    value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales'],
    var_name='Region',
    value_name='Sales_Millions'
)
pref_platform['Region'] = pref_platform['Region'].str.replace('_Sales', '', regex=False)

platform_genre = (
    pref_platform.groupby(['Year', 'Region', 'Platform', 'Genre_Category'], as_index=False)['Sales_Millions']
    .sum()
)

platform_genre['Market_Share_Pct'] = (
    platform_genre.groupby(['Year', 'Region', 'Platform'])['Sales_Millions']
    .transform(lambda x: x / x.sum() * 100 if x.sum() > 0 else 0)
)

avg_platform_genre = (
    platform_genre.groupby(['Platform', 'Region', 'Genre_Category'], as_index=False)['Market_Share_Pct']
    .mean()
)
```

```python
for platform in platforms:
    platform_data = avg_platform_genre[avg_platform_genre['Platform'] == platform].copy()
    pivot = platform_data.pivot(
        index='Region',
        columns='Genre_Category',
        values='Market_Share_Pct'
    )[genre_order].round(1)
    print(f"\n{platform}:")
    display(pivot)
```

```python
# Faceted bar charts - one panel per platform
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

genre_colors = {
    'Action/Competitive': '#e74c3c',
    'Narrative': '#3498db',
    'Strategy/Cognitive': '#2ecc71',
    'Misc': '#f39c12'
}

for idx, platform in enumerate(platforms):
    ax = axes[idx]
    platform_data = avg_platform_genre[avg_platform_genre['Platform'] == platform]
    
    if len(platform_data) > 0:
        pivot = platform_data.pivot(index='Region', columns='Genre_Category', values='Market_Share_Pct')
        pivot = pivot.reindex(index=region_order, columns=genre_order, fill_value=0)
        
        x = np.arange(len(region_order))
        width = 0.2
        
        for i, genre in enumerate(genre_order):
            ax.bar(x + i*width, pivot[genre].values, width, 
                   label=genre, color=genre_colors[genre], alpha=0.85)
        
        ax.set_xlabel('Region', fontweight='bold')
        ax.set_ylabel('Market Share (%)', fontweight='bold')
        ax.set_title(platform, fontweight='bold', fontsize=12)
        ax.set_xticks(x + width * 1.5)
        ax.set_xticklabels(region_order)
        ax.set_ylim(0, 100)
        ax.grid(True, alpha=0.3, axis='y')
        
        if idx == 0:
            ax.legend(loc='upper right', fontsize=9, frameon=False)
    else:
        ax.text(0.5, 0.5, f'No data for {platform}', 
                ha='center', va='center', transform=ax.transAxes)
        ax.set_xticks([])
        ax.set_yticks([])

plt.tight_layout()
plt.suptitle('Average Genre Market Share by Platform and Region (2006–2015)', 
             y=1.00, fontweight='bold', fontsize=14)
plt.show()
```
